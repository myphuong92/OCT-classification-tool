```python
import os
import pandas as pd

# Đường dẫn đến thư mục chứa các file CSV gốc
input_folder = "/content/drive/My Drive/Colab Notebooks/OCT classification/Ensemble/csv"

# Đường dẫn đến thư mục mới
output_folder = "/content/drive/My Drive/Colab Notebooks/OCT classification/Ensemble/csv/output"

# Kiểm tra nếu thư mục đầu ra chưa tồn tại, tạo mới
if not os.path.exists(output_folder):
    os.makedirs(output_folder)
```

```python
import numpy as np
import matplotlib.pyplot as plt

# Hàm Exponential
def default_exponential_function(x):
    return 1 - np.exp(-((x-1)**2)/2.0) # Default Exponential Function

# Tạo dữ liệu đầu vào
x = np.linspace(0, 1, 100)  # Tạo 100 điểm từ 0 đến 10

# Tính toán giá trị hàm Exponential tương ứng với mỗi điểm x
y = default_exponential_function(x)

# Vẽ đồ thị
plt.plot(x, y, label='Default Exponential Function')
plt.xlabel('x')
plt.ylabel('Default Exponential Function Value')
plt.title('Plot of Default Exponential Function')
plt.legend()
plt.grid(True)
plt.show()
```

## **Vẽ đồ thị hàm phân phối: Gompertz**

```python
import numpy as np
import matplotlib.pyplot as plt

# Hàm Gompertz
def gompertz_function(x, eta, b):
    return 1 - np.exp(-eta * np.exp(-b *x))

def weighted_gompertz_function(x, s, l):
    return 1 - (1+(s * (np.exp(l*x)))/(1+l*s))* np.exp(-s * (np.exp(l*x)))

# Thiết lập giá trị của tham số
s = 1.8324 #2.97 #1.87
l = 0.5637 # 0.82275#82275 #

# Thiết lập giá trị của tham số
eta = 1.846
b = 17.4
eta2 = 1
b2 = 14
# Tạo dữ liệu đầu vào
x = np.linspace(0, 1, 100)  # Tạo 100 điểm từ 0 đến 10

# Tính toán giá trị hàm Gompertz tương ứng với mỗi điểm x
y = gompertz_function(x, 1, 2)
y1 = gompertz_function(x, eta2, b2)
y2 = weighted_gompertz_function(x, s, l)
y3 = weighted_gompertz_function(x, s, l) * gompertz_function(x, eta2, b2)
# Vẽ đồ thị
plt.plot(x, y, label='Gompertz Function (eta = 1, b = 2)')
plt.plot(x, y1, label='Gompertz Function (eta = 1, b = 14)')

plt.plot(x, y2, label='weighted_gompertz_function (s = 1.97, l = 0.82275)')

plt.plot(x, y3, label='ave weighted_gompertz_function (s = 1.97, l = 0.82275)')

plt.xlabel('x')
plt.ylabel('Gompertz Function Value')
plt.title('Plot of Gompertz Function')
plt.legend()
plt.grid(True)
plt.show()
```

# **Vẽ đồ thị hàm phân phối: Shift Gompertz**

```python
import numpy as np
import matplotlib.pyplot as plt

# Hàm Shift Gompertz
def shift_gompertz_function(x, eta, b):
    return (1- np.exp(-b * x)) * np.exp(-eta * np.exp(-b * x)) # Shitf Gompertz

# Thiết lập giá trị của tham số
eta = 1.132
b = 3.423
# Tạo dữ liệu đầu vào
x = np.linspace(0, 1, 100)  # Tạo 100 điểm từ 0 đến 10

# Tính toán giá trị hàm Gompertz tương ứng với mỗi điểm x
y = shift_gompertz_function(x, eta, b)

# Vẽ đồ thị
plt.plot(x, y, label='Shift Gompertz Function (eta = -0.0932; b = 2.0829)')
plt.xlabel('x')
plt.ylabel('Shift Gompertz Function Value')
plt.title('Plot of Shift Gompertz Function')
plt.legend()
plt.grid(True)
plt.show()
```

# **Vẽ đồ thị hàm phân phối: Weighted Gompertz**

```python
import numpy as np
import matplotlib.pyplot as plt

# Hàm Weighted Gompertz
def weighted_gompertz_function(x, s, l):
    return 1 - (1+(s * (np.exp(l*x)-1))/(1+l*s))* np.exp(-s * (np.exp(l*x)-1))

# Thiết lập giá trị của tham số
s = 1.97 #1.87
l = 0.82275#82275 #

# Tạo dữ liệu đầu vào
x = np.linspace(0, 1, 100)  # Tạo 100 điểm từ 0 đến 10

# Tính toán giá trị hàm Gompertz tương ứng với mỗi điểm x
y = weighted_gompertz_function(x, s, l)

# Vẽ đồ thị
plt.plot(x, y, label='weighted_gompertz_function (s = 1.97, l = 0.82275)')
plt.xlabel('x')
plt.ylabel('Weighted Gompertz Function Value')
plt.title('Plot of Weighted Gompertz Function')
plt.legend()
plt.grid(True)
plt.show()
```

# **Vẽ đồ thị hàm phân phối: hyperbolic tangent function**

```python
import numpy as np
import matplotlib.pyplot as plt

# Hàm hyperbolic tangent
def tangent_function(x, eta, b):
    return (1 - np.tanh(((x-1)**2)/2)) * (1 - np.exp(-((x-1)**2)/2.0))

# Thiết lập giá trị của tham số
eta = 0.5
b = -0.4

# Tạo dữ liệu đầu vào
x = np.linspace(-2, 1, 100)  # Tạo 100 điểm từ 0 đến 10

# Tính toán giá trị hàm Gompertz tương ứng với mỗi điểm x
y = tangent_function(x, eta, b)

# Vẽ đồ thị
plt.plot(x, y, label='hyperbolic tangent Function')
plt.xlabel('x')
plt.ylabel('hyperbolic tangent Value')
plt.title('Plot of hyperbolic tangent Function')
plt.legend()
plt.grid(True)
plt.show()
```
